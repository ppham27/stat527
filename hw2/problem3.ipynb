```python
import pickle
import numpy as np
from sklearn import model_selection
from statsmodels.nonparametric import kernel_regression

np.set_printoptions(suppress=True)

kernel_regression.kernel_func['box'] = (
    lambda h, Xi, x: np.where(np.abs(Xi - x) <= h, 0.5/h, 0))

def make_data(n=100, seed=2020):
    np.random.seed(seed)
    x = np.random.uniform(-1, 1, size=[n])
    epsilon = np.random.normal(size=x.shape)
    order = np.argsort(x)
    return x[order], epsilon[order]

def f_sin(x):
    return np.sin(x * np.pi)

def f_poly(x):
    x3 = x * x * x
    return 2 * x + x3 - 6 * x * x3

def f_inv(x):
    return 1. / (1. + 25 * x * x)
```

```python
x, epsilon = make_data()
```

```python
def experiment(x, epsilon, f, min_h, max_h, num_folds, ckertype):
    y = f(x) + epsilon
    h_candidates = np.linspace(min_h, max_h, 100)
    # Cross validation
    cv_errors = np.zeros_like(h_candidates)    
    kfold = model_selection.KFold(num_folds, shuffle=True)
    for train_split, test_split in kfold.split(x):
        x_train, x_test, y_train, y_test = (
            x[train_split], x[test_split], y[train_split], y[test_split])
        for i, h in enumerate(h_candidates):
            model = kernel_regression.KernelReg(
                y_train, x_train, var_type=['c'], reg_type='lc', bw=[h], ckertype=ckertype)
            y_pred, _ = model.fit(x_test)
            cv_errors[i] += np.mean(np.square(y_pred - y_test))            
    cv_errors /= num_folds
    # Oracle
    oracle_errors = np.zeros_like(h_candidates)
    for i, h in enumerate(h_candidates):        
        model = kernel_regression.KernelReg(
            y, x, var_type=['c'], reg_type='lc', bw=[h], ckertype=ckertype)        
        y_pred, _ = model.fit(x)
        oracle_errors[i] = np.mean(np.square(y_pred - f(x)))
    return h_candidates[np.nanargmin(cv_errors)], h_candidates[np.nanargmin(oracle_errors)]

experiment(x, epsilon, f_poly, 0.01, 1., 10, 'box')
```

```python
import collections

import collections
random_fold_experiments = collections.defaultdict(
    lambda: collections.defaultdict(
        lambda: {'oracle': np.nan, 'simulations': collections.defaultdict(list)}))

for i in range(100):
    for name, f in [('sin', f_sin), ('poly', f_poly), ('inv', f_inv)]:
        for kernel in ['box', 'gaussian']:
            for k in [2, 5, 10]:
                h, h_oracle = experiment(x, epsilon, f, 0.01, 1., k, kernel)
                random_fold_experiments[name][kernel]['oracle'] = h_oracle
                random_fold_experiments[name][kernel]['simulations'][k].append(h)
```
